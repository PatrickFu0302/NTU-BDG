```python
import pandas as pd
import numpy as np
import requests
```

# Get power meter data

<img src="picture/powerMeterData.jpg" width="60%">

```python
#可以從chrome的DevTool裡面的network觀看, data form裡面的參數作為payload進行post
#參考連結1: https://stackoverflow.com/questions/56183029/how-to-read-data-from-from-web-link-with-drop-down-fields-using-python
#參考連結2: https://stackoverflow.com/questions/54789508/using-python-requests-module-with-dropdown-options

payload = {'ctg': 'all',
           'dt1': '2019/08/09',
           'ok': '查詢'}

r = requests.post('http://140.112.166.97/power/fn4/report2.aspx', data = payload)
print(r.text)
```

```python
tables = pd.read_html(r.text)
tables[0]
```

```python
tables[1]
```

```python
payload = {'SYearDDL': '1081',
           'BuildingDDL': '共同',
           'RoomDDL': '共102',
           'SelectButton': '查詢'}

#要注意把url的https改成http, read_html才能讀取
r = requests.get('http://gra206.aca.ntu.edu.tw/classrm/index.php/acarm/webcr-use-new', params = payload)
print(r.text)
```
